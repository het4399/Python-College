```python
income=[5,9,6,6,10,7,6,4,4,5,6,4]
expenses=[6,6,8,3,6,9,7,8,6,6,4,8]
time=range(1,13)
fig,ax=plt.subplots(figsize=(8,8))
ax.plot(time,income,color='g')
ax.plot(time,expenses,color='b')
ax.fill_between(time,income,expenses,where=[i>j for i,j in zip(income,expenses)],color='g',alpha=0.4,label="positive",interpolate="True")
ax.fill_between(time,income,expenses,where=[i<=j for i,j in zip(income,expenses)],color='b',alpha=0.4,label="negative",interpolate="True")

```

```python
df=pd.read_csv("Police_Department_Incidents_-_Previous_Year__2016_.csv")
df=df.iloc[0:200,:]
map=folium.Map(location=[36.77,-122.494],zoom_start=10)
incidents=folium.FeatureGroup().add_to(map)
for lat,len,cat in zip(df.Y,df.X,df.Category):
    folium.CircleMarker(location=[lat,len],radius=5,color='yellow',fill=True,fill_color='blue',fill_opacity=.6,popup=cat).add_to(incidents)
map.save("CRIME.HTML")
```

```python
import matplotlib.pyplot as plt
import networkx as nx
G=nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,1)
nx.draw(G,with_labels=True)
plt.show()
```

```python
G=nx.DiGraph()
G.add_nodes_from(['A','B','C','D'])
G.add_edges_from([('A','B'),('A',"C"),('A','D'),('C','D'),('D','C')])
nx.draw(G,with_labels=True)
plt.show()
```

```python
# by a company to analyse the social network of 
# company has provided following data

# there are 5 employee in the company which has a following relationship

# 1).Employee 1 is friend with employee 2and3
# 2)Employee 2 is friend with employee 4  
# 3)Emp 3 is friend with EMp5 
# 4)To  create the grap 
a=nx.DiGraph()
a.add_nodes_from([1,2,3,4,5])
a.add_edges_from([(1,2),(1,3),(2,4),(3,5)])
labels={1:'Employee1',2:'Employee2',3:'Employee3',4:'Employee4',5:'Employee5'}
nx.draw(a,with_labels=True,labels=labels)
plt.show()
```
